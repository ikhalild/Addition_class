## üéØ **Nettoyage des colonnes num√©riques**
Nous allons convertir les colonnes de prix, rating et nombre d'avis en float. Nous allons aussi enlever le signe % de la colonne discount_percentage

```python
# Cr√©ons d'abord une copie de notre jeu de donn√©es
df1 = df.copy()
```

```python
# Dressons le tableau des effectifs de la colonne rating pour voir pourquoi c'est de type object
df1['rating'].value_counts()
```

```python
# On a remarqu√© la pr√©sence de | dans la colonne rating
# Essayons de le remplacer par une value manquant (on peut aussi d√©cider de le remplacer avec la m√©diane ou la moyenne)
df1['rating'] = df1['rating'].replace('|', np.nan)
```

```python
# Enlevons le symbol de devise et le d√©limitateur de millier
df1['discounted_price'] = df1['discounted_price'].str.replace("‚Çπ", "").str.replace(",", "").astype(float)
df1['actual_price'] = df1['actual_price'].str.replace("‚Çπ", "").str.replace(",", "").astype(float)
df1['rating'] = df1['rating'].str.replace(",", "").astype(float)
df1['rating_count'] = df1['rating_count'].str.replace(",", "").astype(float)
```

```python
# Changeons le type de donn√©e et ramenons le pourcentage en proportion (entre 0 et 1) dans discount_Percentage
df1['discount_percentage'] = df1['discount_percentage'].str.replace('%','').astype('float64')
df1['discount_percentage'] = df1['discount_percentage'] / 100
```

```python
# Affichons les colonne modifi√©es
df1[['discounted_price', 'actual_price', 'rating', 'rating_count', 'discount_percentage']].head()
```

```python
# Avant d'entamer les probl√®mes, faisons une nouvelle copie de notre jeu de donn√©es
df2 = df1.copy()
```

# **Probl√®me 1**

**Quel est le vrai prix sans discount du produit le moins cher parmi les mieux not√©s ?**

```python
# Que demande ce probl√®me ?
# On cherche le vrai prix dans discount (c'est-√†-dire la colonne actual_price) du produit le moins cher (colonne discounted price)
# parmi les mieux not√©s (colonne rating)

# Pour ce faire il faut :
# 1. Trouver la meilleure note
# 2. Filtrer les produits ayant cette note
# 3. Trouver le prix le plus bas parmi ceux-l√†
# 4. Garder uniquement les lignes qui ont ce prix
# 5. Afficher le r√©sultat
```

```python
# 1. Trouver la meilleure note
meilleur_note = df2['rating'].max()
meilleur_note
```

```python
# 2. Filtrer les produits ayant cette note
high_rating = df2[df2['rating'] == meilleur_note]
high_rating
```

```python
# 3. Trouver le prix le plus bas parmi ceux-l√†
min_price = high_rating['discounted_price'].min()
min_price
```

```python
# 4. Garder uniquement les lignes qui ont ce prix
result = high_rating[high_rating['discounted_price'] == min_price]
result
```

```python
# 5. Afficher le r√©sultat
result[['actual_price']]
```

# **Probl√®me 2**

**Quel est le vrai prix sans discount du produit le mieux not√© parmi les moins chers ?**

```python
# Le raisonnement est le m√©me que dans le probl√®me 1
```

```python
# 1. Trouver le prix le plus bas globalement
min_price = df2['discounted_price'].min()
```

```python
# 2. Filtrer les produits ayant ce prix
cheapest_products = df2[df2['discounted_price'] == min_price]
```

```python
# 3. Parmi eux, trouver ceux ayant la meilleure note
max_rating = cheapest_products['rating'].max()
```

```python
# 4. S√©lectionner les produits avec ce rating maximal
result = cheapest_products[cheapest_products['rating'] == max_rating]
```

```python
# 5. Afficher le vrai prix sans discount
result['actual_price']
```

# **Probl√®me 3**

**Comment calculer et afficher le prix final TTC (avec 18% de TVA sur le prix remis√©)**

```python
# On commence par cr√©er une nouvelle colonne contant la TVA
df2['TVA'] = df2['discounted_price'] * 0.18

# Ensuite une nouvelle colonne pour le prix Tout Taxe Compris (qui est la somme entre la TVA et le prix apr√®s r√©duction)
df2['Prix_TTC'] = df2['discounted_price'] + df2['TVA']

# On affiche le 5 premi√®re ligne du r√©sultat
df2[['discounted_price', 'TVA', 'Prix_TTC']].head()
```

```python
# On calcule le prix TTC moyen
df2['Prix_TTC'].mean()
```

```python
# On regroupe les lignes par cat√©gorie
df2.groupby('category')
```

```python
# On calcule le prix TTC moyen pour chaque cat√©gorie
df2.groupby('category')['Prix_TTC'].mean()
```

```python
# On affiche les 6 premi√®res cat√©gories
df2.groupby('category')['Prix_TTC'].mean().sort_values(ascending=False).head(6)
```

**Renormer la colonne "discounted_price"**

```python
# Faisons d'abord une nouvelle copie de notre jeu de donn√©es avant de le modifier
df3 = df2.copy()
```

```python
# Maintenant on renorme la colonne
df3 = df3.rename(columns={"discounted_price": "prix_remis√©"})
```

# **Quiz 1 : Quel est le vrai prix sans discount du produit le moins cher en termes de prix TTC parmi les mieux not√©s ?**

```python
# Que demande ce quiz ?
# C'est la m√™me question que le probl√®me 1 √† l'exception qu'on remplace discounted price par prix TTC
# On cherche le vrai prix dans discount (c'est-√†-dire la colonne actual_price) du produit le moins cher (colonne discounted price)
# parmi les mieux not√©s (colonne rating)

# Pour ce faire il faut :
# 1. Trouver la meilleure note
# 2. Filtrer les produits ayant cette note
# 3. Trouver le prix TTC le plus bas parmi ceux-l√†
# 4. Garder uniquement les lignes qui ont ce prix
# 5. Afficher le r√©sultat
```

```python
# 1. Trouver la meilleure note
meilleur_note = df3['rating'].max()

# 2. Filtrer les produits ayant cette note
high_rating = df3[df3['rating'] == meilleur_note]

# 3. Trouver le prix le plus bas parmi ceux-l√†
min_price = high_rating['Prix_TTC'].min() # Attention, on avait renomm√© la colonne, ce n'est plus discounted_prix

# 4. Garder uniquement les lignes qui ont ce prix
result = high_rating[high_rating['Prix_TTC'] == min_price]

# 5. Afficher le r√©sultat
result[['actual_price']]
```

# **Quiz 2 : Quel est le prix TTC du produit le mieux not√© parmi les moins chers en termes de prix remis√© ?**

```python
# 1. Trouver le prix remis√© le plus bas globalement
min_price = df3['prix_remis√©'].min()

# 2. Filtrer les produits ayant ce prix
cheapest_products = df3[df3['prix_remis√©'] == min_price]

# 3. Parmi eux, trouver ceux ayant la meilleure note
max_rating = cheapest_products['rating'].max()

# 4. S√©lectionner les produits avec ce rating maximal
result = cheapest_products[cheapest_products['rating'] == max_rating]

# 5. Afficher le prix TTC
result['Prix_TTC']
```
